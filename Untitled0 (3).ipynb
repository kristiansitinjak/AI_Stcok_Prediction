# **DATA COLECTION**

```python
import pandas as pd
import numpy as np
import os
import joblib
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

tickers = ["BBRI.JK", "GOTO.JK", "ANTM.JK", "ADRO.JK", "TLKM.JK",
           "MEDC.JK", "BMRI.JK", "ASII.JK", "BUKA.JK", "BBNI.JK"]

folder_path = '/content/drive/MyDrive/ML_SahamPredictionTask'
model_output_path = os.path.join(folder_path, "Models")
if not os.path.exists(model_output_path):
    os.makedirs(model_output_path)

processed_dfs = {}
```

```python
import yfinance as yf
import os

tickers = ["BBRI.JK", "GOTO.JK", "ANTM.JK", "ADRO.JK", "TLKM.JK",
           "MEDC.JK", "BMRI.JK", "ASII.JK", "BUKA.JK", "BBNI.JK"]

folder_path = '/content/drive/MyDrive/ML_SahamPredictionTask'
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

for ticker in tickers:
    try:
        print(f"Mengunduh {ticker}...")
        data = yf.download(ticker, start="2020-01-01", end="2026-02-23", auto_adjust=True, timeout=20)

        if not data.empty:
            if isinstance(data.columns, pd.MultiIndex):
                data.columns = data.columns.get_level_values(0)

            file_name = f"{folder_path}/{ticker}_historical_data.csv"
            data.to_csv(file_name)
            print(f"{ticker} Berhasil disimpan.")
        else:
            print(f"{ticker} Kosong, cek koneksi/ticker.")

    except Exception as e:
        print(f"Gagal mengunduh {ticker}: {e}")

print("\n--- Download Selesai! Silakan lanjut ke Step 1 ---")
```

```python
print("Preprocessing & Feature Engineering")
for ticker in tickers:
    try:
        raw_file = f"{folder_path}/{ticker}_historical_data.csv"
        df = pd.read_csv(raw_file, index_col=0, parse_dates=True)

        # Buat Indikator
        df['MA5'] = df['Close'].rolling(window=5).mean()
        df['MA20'] = df['Close'].rolling(window=20).mean()
        df['Daily_Return'] = df['Close'].pct_change()
        df.dropna(inplace=True)

        # Buat Target (Harga Besok)
        df['Target_Next_Day'] = df['Close'].shift(-1)
        df_final = df.dropna()

        processed_dfs[ticker] = df_final
        print(f"Preprocessing {ticker} Selesai.")
    except Exception as e:
        print(f"‚ùå Gagal Preprocessing {ticker}: {e}")
```

# **Scaling & Splitting**

```python
print("Scaling & Splitting")
scaled_data = {} # Simpan hasil scaling

for ticker in tickers:
    if ticker in processed_dfs:
        df_final = processed_dfs[ticker]
        features = ['Open', 'High', 'Low', 'Close', 'Volume', 'MA5', 'MA20']

        X = df_final[features]
        y = df_final['Target_Next_Day'].values.reshape(-1, 1)

        scaler_X = MinMaxScaler(feature_range=(0, 1))
        scaler_y = MinMaxScaler(feature_range=(0, 1))

        X_scaled = scaler_X.fit_transform(X)
        y_scaled = scaler_y.fit_transform(y)

        train_size = int(len(X_scaled) * 0.8)
        X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
        y_train, y_test = y_scaled[:train_size], y_scaled[train_size:]

        scaled_data[ticker] = {
            'X_train': X_train, 'X_test': X_test,
            'y_train': y_train, 'y_test': y_test,
            'scaler_X': scaler_X, 'scaler_y': scaler_y
        }
        print(f"Scaling {ticker} Selesai.")
```

# **Melatih Model & Simpan**

```python
print("--- Step 3: Melatih Model & Simpan ---")
for ticker in tickers:
    if ticker in scaled_data:
        sd = scaled_data[ticker]

        # Inisialisasi & Training
        model = RandomForestRegressor(n_estimators=100, random_state=42)
        model.fit(sd['X_train'], sd['y_train'].ravel())

        # Simpan Model & Scaler dengan nama spesifik
        joblib.dump(model, f"{model_output_path}/model_{ticker}.pkl")
        joblib.dump(sd['scaler_X'], f"{model_output_path}/scaler_x_{ticker}.pkl")
        joblib.dump(sd['scaler_y'], f"{model_output_path}/scaler_y_{ticker}.pkl")

        # Simpan model di memory untuk Step 4
        scaled_data[ticker]['model'] = model
        print(f"Model {ticker} Berhasil Dilatih & Disimpan.")
```

```python
print("Visualisasi Hasil Prediksi")
for ticker in tickers:
    if ticker in scaled_data:
        sd = scaled_data[ticker]
        model = sd['model']

        # Prediksi pada data uji
        y_pred_scaled = model.predict(sd['X_test'])
        y_test_actual = sd['scaler_y'].inverse_transform(sd['y_test'])
        y_pred_actual = sd['scaler_y'].inverse_transform(y_pred_scaled.reshape(-1, 1))

        accuracy = r2_score(y_test_actual, y_pred_actual)
        print(f"Akurasi {ticker} (R2 Score): {accuracy:.4f}")

        # Plotting
        plt.figure(figsize=(10, 4))
        plt.plot(y_test_actual, label='Actual', color='blue')
        plt.plot(y_pred_actual, label='Predicted', color='red', linestyle='--')
        plt.title(f"Prediksi {ticker} (R2: {accuracy:.4f})")
        plt.legend()
        plt.show()
```
